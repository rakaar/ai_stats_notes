```python
# download MNIST datasets
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, transform=torchvision.transforms.ToTensor(), download=True)
# transform is used to convert from PIL image fmt to tensor
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, transform=torchvision.transforms.ToTensor())
```

```python
# load it in tensor form
batch_size = 50
train_loader = torch.utils.data.DataLoader(dataset = train_dataset, batch_size = batch_size, shuffle = True)
test_loader = torch.utils.data.DataLoader(dataset = test_dataset, batch_size = batch_size, shuffle = False)
```

```python
import matplotlib.pyplot as plt
from IPython.display import clear_output

def live_plot(data_dict, figsize=(7,5), title=''):
    clear_output(wait=True)
    plt.figure(figsize=figsize)
    for label, data in data_dict.items():
        plt.plot(data, label=label)
    plt.title(title)
    plt.grid(True)
    plt.xlabel('batch')
    plt.legend(loc='center left') # Change the location based on your preference
    plt.show()
```

```python
device = torch.device('cuda')
```

```python
class MyNeuralNet(torch.nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        # input to weight matrix
        super(MyNeuralNet, self).__init__() # to run it on cuda
        # wrapping in parameters help u access them using model.parameters()
        self.W1 = torch.nn.Parameter(torch.randn(hidden_size, input_size) * torch.sqrt(torch.tensor(2. / input_size)))
        self.b1 = torch.nn.Parameter(torch.rand(hidden_size))
        self.W2 = torch.nn.Parameter(torch.randn(output_size, hidden_size) * torch.sqrt(torch.tensor(2. / hidden_size)))
        self.b2 = torch.nn.Parameter(torch.rand(output_size))


    def forward(self, input_vec):
        input_vec = input_vec.to(device)
        L1_output = self.W1 @ input_vec + self.b1
        L1_output = torch.nn.functional.relu(L1_output)
        # print('L1 act ',L1_output)
        L2_output = self.W2 @ L1_output + self.b2
        return L2_output
```

```python
model = MyNeuralNet(784, 1000, 10)
model = model.to(device)
```

```python
learning_rate = 0.01
```

```python
losses = {'cross_entropy_loss': []}
for epoch in range(1):
    for idx, (image_tensor, image_label) in enumerate(train_loader):
        # print(idx+1)
        image_tensor_reshape = image_tensor.reshape(-1, 784).to(device)
        cross_entropy_loss = 0
        for b in range(batch_size):
            image_in_a_batch = image_tensor_reshape[b].to(device)
            image_in_batch_label = image_label[b].to(device)
    
            output_of_single_example_in_batch = model.forward(image_in_a_batch)
            # print(output_of_single_example_in_batch)
            # print(image_in_batch_label)
            one_hot_output = torch.zeros_like(output_of_single_example_in_batch)
            one_hot_output[image_in_batch_label] = 1
            # print(one_hot_output)
    
            # softmax of output
            exp_output = torch.exp(output_of_single_example_in_batch)
            exp_output_sum = torch.sum(exp_output)
            softmax_output = exp_output/exp_output_sum
            # print(softmax_output)
            # print(torch.sum(softmax_output))
            # cross entropy loss = - sum(ground truth x log (pred))
            neg_log_pred = -torch.log(softmax_output)
            # print(neg_log_pred)
            cross_entropy_loss += torch.dot(one_hot_output, neg_log_pred)
            # print(cross_entropy_loss)
    
        cross_entropy_loss = cross_entropy_loss/batch_size # mean
        cross_entropy_loss.backward()

        if b > 600:
            learning_rate *= 0.1
            learning_rate = max(0.00001, learning_rate)
        # update the weights with zero grad, (no need to calculate grad of these weight updates)
        with torch.no_grad():
            model.W1 -= learning_rate*model.W1.grad
            model.W2 -= learning_rate*model.W2.grad

            model.b1 -= learning_rate*model.b1.grad
            model.b2 -= learning_rate*model.b2.grad

            # set the grads to zero
            # Manually zero the gradients after the update
        model.W1.grad.zero_()
        model.b1.grad.zero_()
        model.W2.grad.zero_()
        model.b2.grad.zero_()

        losses['cross_entropy_loss'].append(cross_entropy_loss.item())
            
        # Update the plot
        live_plot(losses, title='Training Loss per batch')
        # print('after a batch Loss = ', cross_entropy_loss.item())  
            # break
        # break
```

```python
# testing the model
with torch.no_grad():
    n_correct = 0
    n_samples = len(test_loader.dataset)

    # print(n_samples)
    for idx, (image, label) in enumerate(test_loader):
        image_tensor_reshape = image.reshape(-1, 784).to(device)
        for b in range(batch_size):
            image_single_eg = image_tensor_reshape[b].to(device)
            label_single_eg = label[b].to(device)

            output_single_eg = model.forward(image_single_eg)
            # print(output_single_eg)
            predicted = torch.argmax(output_single_eg)
            if label_single_eg == predicted:
                n_correct += 1
                
            

print('accuracy = ', 100*n_correct/n_samples)
            
```

# Batch Norm

```python
import torch.nn as nn
```

```python
# write using layers, so that it is easy to do batch norm
class NewNeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(NewNeuralNetwork, self).__init__() # to borrow from nn.Module and run it on cuda
        self.L1 = nn.Linear(input_size, hidden_size)
        self.L2 = nn.ReLU()
        self.L3 = nn.Linear(hidden_size, output_size)
        # batch norm of middle layer
        self.gamma = torch.nn.Parameter(torch.ones(hidden_size))
        self.beta = torch.nn.Parameter(torch.zeros(hidden_size))
        self.is_training = True

    def forward(self, x):
        out = self.L1(x)
        # batch norm before activation(L2) here
        # mean of layer 1
        if self.is_training:
            mean_L1 = torch.mean(out, dim=0)
            var_L1 = torch.var(out, dim=0)
            out = (out - mean_L1)/torch.sqrt(var_L1 + 1e-5)
            out = self.gamma*out + self.beta
            
        out = self.L2(out)
        out = self.L3(out)

        
        return out
        
```

```python
input_size = 784
hidden_size = 500
output_size = 10
```

```python
model = NewNeuralNetwork(input_size, hidden_size, output_size)
model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
learning_rate = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)
```

```python
# Adam Optimizer of Pytorch and SGD custom
losses2 = {'cross_entropy_loss': []}
# model.is_training  = False
n_epochs = 1
for epoch in range(n_epochs):
    for i, (images, labels) in enumerate(train_loader):
        images_reshaped = images.reshape(-1, 784).to(device)
        labels = labels.to(device)

        predicted = model.forward(images_reshaped)
        loss = criterion(predicted, labels)

        loss.backward()
        with torch.no_grad():
            # - default pytroch 
            optimizer.step()
            # - after a batch: Stochastic Gradient Descent(SGD)
            # for param in model.parameters():
            #     param -= learning_rate*param.grad
            optimizer.zero_grad()
        losses2['cross_entropy_loss'].append(loss.item())
        live_plot(losses2, title='Training Loss per batch')
```

```python
# RMS prop
learning_rate = 0.001
alpha = 0.9
epsilon = 1e-8

square_grad = {name: torch.zeros_like(param) for name, param in model.named_parameters()}

losses3 = {'rms_prop': []}
# model.is_training  = False
n_epochs = 1
for epoch in range(n_epochs):
    for i, (images, labels) in enumerate(train_loader):
        images_reshaped = images.reshape(-1, 784).to(device)
        labels = labels.to(device)

        predicted = model.forward(images_reshaped)
        loss = criterion(predicted, labels)

        loss.backward()
        with torch.no_grad():
            # RMS prop
            for name, param in model.named_parameters():
                # print('here')
                square_grad[name] = alpha*square_grad[name] + (1 - alpha)*(param.grad**2)
                param -= (learning_rate * (param.grad))/(torch.sqrt(square_grad[name]) + epsilon)

            for _,param in model.named_parameters():
                param.grad.zero_()
            # - default pytroch 
            # optimizer.step()
            # - after a batch: Stochastic Gradient Descent(SGD)
            # for param in model.parameters():
            #     param -= learning_rate*param.grad
            # optimizer.zero_grad()
        losses3['rms_prop'].append(loss.item())
        live_plot(losses3, title='Training Loss per batch')
```

```python
# Adam Optimizer
beta1 = 0.9
beta2 = 0.999
epsilon = 1e-8
t = 1
learning_rate = 0.01

square_grad = {name: torch.zeros_like(param) for name, param in model.named_parameters()}
just_grad = {name: torch.zeros_like(param) for name, param in model.named_parameters()}

losses4 = {'adam': []}
# model.is_training  = False
n_epochs = 1
for epoch in range(n_epochs):
    for i, (images, labels) in enumerate(train_loader):
        images_reshaped = images.reshape(-1, 784).to(device)
        labels = labels.to(device)

        predicted = model.forward(images_reshaped)
        loss = criterion(predicted, labels)

        loss.backward()
        with torch.no_grad():
            # Adam
            for name, param in model.named_parameters():
                # print('here')
                just_grad[name] = beta1*(just_grad[name]) + (1-beta1)*(param.grad)
                square_grad[name] = beta2*(square_grad[name]) + (1-beta2)*(param.grad**2)

                # corrected bias due to zero initialization
                m_corrected = just_grad[name]/(1 - beta1**t)
                v_corrected = square_grad[name]/(1-beta2**t)
                
                param -= (learning_rate*m_corrected) / (torch.sqrt(v_corrected) + epsilon) 
            
            for _,param in model.named_parameters():
                param.grad.zero_()

            # optimizer.step()
            # optimizer.zero_grad()
            # - default pytroch 
            # optimizer.step()
            # - after a batch: Stochastic Gradient Descent(SGD)
            # for param in model.parameters():
            #     param -= learning_rate*param.grad
            # optimizer.zero_grad()
        losses4['adam'].append(loss.item())
        t = t + 1 # time update
        live_plot(losses4, title='Training Loss per batch')
```

```python
# My code
model.is_training = False
with torch.no_grad():
    n_correct = 0
    n_samples = len(test_loader.dataset)

    # print(n_samples)
    for idx, (images, labels) in enumerate(test_loader):
        image_tensor_reshape = images.reshape(-1, 784).to(device)
        labels = labels.to(device)
        
        outputs = model.forward(image_tensor_reshape)
        _, predicted = torch.max(outputs,1)
       
        n_correct += (predicted == labels).sum().item()
        # print((predicted == labels).sum().item())
        
            

print('accuracy = ', 100*n_correct/n_samples)
```
